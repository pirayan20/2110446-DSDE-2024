### 3.1 pipeline for features

```python
from sklearn.base import BaseEstimator, TransformerMixin

class InsAdjuster(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X_transformed = X.copy()
        X[X < 0.01] = 0

        return X_transformed
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

key = ['key']
ins = ['instrumentalness']

key_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='constant', fill_value=1)),
    ('one-hot', OneHotEncoder(handle_unknown='ignore', sparse=False))
])

ins_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='constant', fill_value=0)),
    ('adjust', InsAdjuster())
])
```

### 3.2 ColumnTransformer

```python
from sklearn.compose import ColumnTransformer

col_trans = ColumnTransformer(transformers=[
    ('key-pipeline', key_pipeline, key),
    ('ins-pipeline', ins_pipeline, ins)
], n_jobs=-1)
```

### 3.3 Add a Model to the pipeline

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier(max_depth=10, min_samples_leaf=10, criterion='gini')

model_pipeline = Pipeline(steps=[
    ('col_trans', col_trans),
    ('model', dtree)
])
```

```python
from sklearn import set_config

set_config(display='diagram')
display(model_pipeline)
```

## 4.	Train/Test split

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# 4.1
y = selected_df.pop("Class")
X = selected_df

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=1234)
```

### Eval without MLflow

```python
model_pipeline.fit(X_train, y_train)
y_pred = model_pipeline.predict(X_test)
print(classification_report(y_test, y_pred, digits=4))
```

## 5.MLflow

```python
# Importing all Libraries
import mlflow
import mlflow.sklearn
from sklearn.metrics import accuracy_score

def eval_metrics(actual, pred):
    accuracy = accuracy_score(actual, pred)
    return accuracy

def train_model(criterion, max_depth, min_samp):

    # Starting the Experiement
    with mlflow.start_run():

        # Model building
        model_pipeline.set_params(model__criterion=criterion, model__max_depth=max_depth, model__min_samples_leaf=min_samp)
        model_pipeline.fit(X_train, y_train)

        y_pred = model_pipeline.predict(X_test)
        (accuracy) = eval_metrics(y_test, y_pred)

        print('Decirion tree (criterion=%s, max_depth=%d):'%(criterion, max_depth))
        print('Accuracy: {:.4f}'.format(accuracy))

        # Logging Parameters
        mlflow.log_param("criterion", criterion)
        mlflow.log_param("max_depth", max_depth)

        # Logging Metrics
        mlflow.log_metric("accuracy", accuracy_score(y_test, y_pred))

        # Model Logging
        mlflow.sklearn.log_model(model_pipeline, 'model')

        return model_pipeline
```

```python
train_model('gini', 10, 10)
```

```python
train_model('entropy', 10, 10)
```

```python
train_model('gini', 5, 2)
```

```python
train_model('entropy',5 , 2)
```

```python
train_model('gini', 4, 1)
```

```python
train_model('entropy', 4, 1)
```

```python
# Load model as a PyFuncModel.
loaded_model = mlflow.pyfunc.load_model(model_uri=f"runs:/{run_id}/model")

# Predict on a Pandas DataFrame.
predicted = loaded_model.predict(pd.DataFrame(X_test))

print(classification_report(y_test, predicted, digits=4))
```
