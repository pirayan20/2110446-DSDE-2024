## 7.	Compare the models [1 point]
* 7.1.	Since there is a single test data set, using McNemar’s test to answer as to whether or not both methods are significantly different.
    -	“Write your answer” Is it significant different?


```python
# I don't think our class cover this one, but better have than not have it right?
```

```python
import numpy as np
from statsmodels.stats.contingency_tables import mcnemar
```

```python
tp_fp = np.zeros((2, 2), dtype=int)

# Construct the contingency table based on predictions and actual outcomes
for predictions, predictions2, actual in zip(predictions, predictions2, y_test):
    if predictions == actual and predictions2 == actual:
        tp_fp[0, 0] += 1  # Both correct
    elif predictions == actual and predictions2 != actual:
        tp_fp[0, 1] += 1  # dtree correct, lgmodel wrong
    elif predictions != actual and predictions2 == actual:
        tp_fp[1, 0] += 1  # dtree wrong, lgmodel correct
    else:
        tp_fp[1, 1] += 1  # Both wrong

# Perform McNemar's test
mcnemar_test_result = mcnemar(tp_fp, exact=True)  # exact=True for small sample sizes

print("Contingency Table:")
print(tp_fp)

print("\nMcNemar's Test Results:")
print(f"Test Statistic: {mcnemar_test_result.statistic}")
print(f"P-Value: {mcnemar_test_result.pvalue}")
```

```python
'''Yes, it is significantly different. The p-value of 3.97988579116416e-22
is much lower than the typical threshold for statistical significance,
which is usually set at 0.05 or 0.01. When the p-value is this small,
it indicates that the observed differences between the
two models are highly unlikely to be due to random chance,
suggesting a significant difference in their performance.
'''
```
