## Compare "Sports" and "Comady", how many days that there are more total daily views of VDO in "Sports" category than in "Comady" category?  **Ans:49**

```python
category = pd.read_json('data/GB_category_id.json')
```

```python
import json
with open('data/GB_category_id.json') as fd:
  cat = json.load(fd)
```

```python
cat_list = []
for each in cat['items']:
  cat_list.append((int(each['id']), each['snippet']['title']))
```

```python
cat_list
```

```python
data_cat = pd.DataFrame(cat_list, columns=["cat_id",'category'])
```

```python
data = data.merge(data_cat, left_on="category_id",right_on="cat_id")
```

```python
data_sports = data[data.category == "Sports"]
data_comedy = data[data.category == "Comedy"]
```

```python
group_sports = data_sports.groupby("trending_date")["views"].sum().reset_index(name="total_sports_views")
group_comedy = data_comedy.groupby("trending_date")["views"].sum().reset_index(name="total_comedy_views")
```

```python
data_final = group_comedy.merge(group_sports, left_on="trending_date", right_on="trending_date")
data_final
```

```python
len(data_final[data_final.total_comedy_views < data_final.total_sports_views])
```
