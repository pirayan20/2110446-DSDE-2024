```python
import numpy as np
import pandas as pd
```

# 1. Load the "ModifiedEdibleMushroom.csv" data from the link below (note: this data set has been preliminarily prepared.) again.

```python
data = pd.read_csv('https://raw.githubusercontent.com/pvateekul/2110446_DSDE_2023s2/main/DataSet/hed2020_dataset.csv')
```

```python
data.head()
```

# 2. Choose edible mushrooms only.

```python
data = data[data.label == 'e']
```

# 3.  Only the variables below have been selected to describe the distinctive characteristics of edible mushrooms
<font color=red> 'cap-color-rate', 'stalk-color-above-ring-rate'

```python
num_col = ['cap-color-rate', 'stalk-color-above-ring-rate']
data = data[num_col]
```

```python
data.isnull().sum()
```

# 4. Provide a proper data preprocessing as follows:

- Fill missing with mean

- Standardize variables with Standard Scaler

```python
data.fillna(data.mean(), inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

# get & fit the scaler
scaler = StandardScaler()
data[num_col] = scaler.fit_transform(data[:])
```

# 5. K-means clustering with 5 clusters and seed = 2020.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=2020).fit(data)
```

```python
data['cluster'] = kmeans.labels_
```

```python
data
```

# 6. Show the centroid of each cluster.

```python
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=num_col)
centroids
```

# 7. Convert the centroid value to the original scale

```python
centroids[num_col] = scaler.inverse_transform(centroids[num_col])
centroids
```
