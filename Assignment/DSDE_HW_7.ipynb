## How many users are there in the database?

```python
cursor = 0
total = 1
cursor, keys = rd.scan(cursor, match='user:*:name')
while cursor > 0:
    cursor, keys = rd.scan(cursor, match='user:*:name')
    total += len(keys)

total
```

## What is the average number of follows per user?

```python
total_follows = 0
cursor = 0

while True:
    cursor, keys = rd.scan(cursor, match='user:*:name')
    for key in keys:
        user_id = key.split(':')[1]
        total_follows += rd.scard(f'user:{user_id}:follows')
    if cursor == 0:
        break

avg = total_follows / total if total > 0 else 0
avg
```

## How many users follows between 5-10 users?

```python
users_following_5_to_10 = 0

cursor = 0

while True:
    cursor, keys = rd.scan(cursor, match='user:*:name')
    for key in keys:
        user_id = key.split(':')[1]
        follows_count = rd.scard(f'user:{user_id}:follows')
        if 5 <= follows_count <= 10:
            users_following_5_to_10 += 1
    if cursor == 0:
        break

users_following_5_to_10
```

## Which account has the most followers?

```python
max_followers = 0
user_with_max_followers_id = None

cursor = 0

while True:
    cursor, keys = rd.scan(cursor, match='user:*:name')
    for key in keys:
        user_id = key.split(':')[1]
        followers_count = rd.scard(f'user:{user_id}:followed_by')
        if followers_count > max_followers:
            max_followers = followers_count
            user_with_max_followers_id = user_id
    if cursor == 0:
        break

user_with_max_followers_id
```
